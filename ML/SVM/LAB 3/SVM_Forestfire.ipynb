```python
forest_data11=forest_data.drop('month',axis=1)
```

```python
forest_data12=forest_data11.drop('day',axis=1)
```

```python
forest_data12.head()
```

```python
forest_data12.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
forest_data12.iloc[:, 28] = labelencoder.fit_transform(forest_data12.iloc[:,28])
```

```python
forest_data12.head()
```

```python
X = forest_data12.iloc[: ,0:28]
```

```python
Y = forest_data12.iloc[: ,28]
```

```python
Y
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
clf = SVC()
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
clf = SVC(C= 15, gamma = 50)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
