```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
# from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
company=pd.read_csv(r"C:\Users\rajni\Desktop\Ml\LAB 2\Company_Data.csv")
```

```python
company.head()
```

```python
company.describe()
```

```python
company1=pd.get_dummies(company,columns=['Urban','US','ShelveLoc'])
```

```python
company1.head()
```

```python
company1["NewSale"]="<=7.490000"
company1.loc[company1["Sales"]>=7.490000,"NewSale"]="HIGH"
company1.loc[company1["Sales"]<=7.490000,"NewSale"]="LOW"
```

```python
company1
```

```python
company11=company1.drop('Sales',axis=1)
```

```python
company11.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
company11.iloc[:,14] = labelencoder.fit_transform(company11.iloc[:,14])
```

```python
company11
```

```python
X = company11.iloc[:,0:14]
Y = company11.iloc[:,14]
```

```python
Y
```

```python
X
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2,random_state=40)
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
tree.plot_tree(model);
```

```python
preds = model.predict(x_test) # predicting on test data set 
pd.Series(preds).value_counts() # getting the count of each category 
```

```python
preds
```

```python
pd.crosstab(y_test,preds) # getting the 2 way table to understand the correct and wrong predictions
```

```python
# Accuracy 
np.mean(preds==y_test)
```
